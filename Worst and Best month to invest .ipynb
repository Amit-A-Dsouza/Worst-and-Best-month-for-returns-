### Worst and Best month for returns using S&P500 Data 1927 to 2020

```python
## Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Loading Data

sp500= pd.read_csv("SPX.csv")

##Set Date as Index
sp500=sp500.set_index(pd.DatetimeIndex(sp500["Date"]))
sp500
```

```python
## Create list to store the Returns of each day in that month

Jan = []
Feb = []
Mar = []
Apr = []
May = []
June = []
July = []
Aug = []
Sept = []
Oct = []
Nov = []
Dec = []

## Creating functions to get returns for each month

def get_returns(month, returns):
    if month=='01':
        Jan.append(float(returns))
    elif month== '02':
        Feb.append(float(returns))
    elif month== '03':
        Mar.append(float(returns))
    elif month== '04':
        Apr.append(float(returns))
    elif month== '05':
        May.append(float(returns))
    elif month== '06':
        June.append(float(returns))
    elif month== '07':
        July.append(float(returns))
    elif month== '08':
        Aug.append(float(returns))
    elif month== '09':
        Sept.append(float(returns))
    elif month== '10':
        Oct.append(float(returns))
    elif month== '11':
        Nov.append(float(returns))
    elif month== '12':
        Dec.append(float(returns))
    else:
        print("Something is Wrong with the input for month ")
```

```python
## Calculate Daily simple returns

DSR= sp500["Close"].pct_change(1)
sp500["DSR"]= DSR

sp500

```

```python
### Remove first Row of the data from the dataset

sp500= sp500[1:]

sp500
```

```python
## Create a loop to gather the daily  simple returns of each month from the dataset and append them to list

for i in range (0, len(sp500)):
    sp500_date= str(sp500.index[i])
    sp500_returns= sp500['DSR'][i]
    sp500_month= sp500_date.split('-')[1]
    
    ## add returns to that list corresponding to that month 
    
    get_returns (sp500_month, sp500_returns)
```

```python
## Create a function to Average the return for each month 

def AVG(month):
    return [sum(month)/len(month)]
```

```python
## Create a new dataframe
sp500_AVG = pd.DataFrame()

## Get the average return for each month and add the values under a new column

sp500_AVG['AVG']= AVG(Jan)+ AVG(Feb)+AVG(Mar)+ AVG(Apr)+ AVG(May)+ AVG(June) +AVG(July)+ AVG(Aug)+ AVG(Sept)+ AVG(Oct)+ AVG(Nov)+ AVG(Dec)

## Set the index to be corresponding integer value of the month

sp500_AVG= sp500_AVG.set_index(sp500_AVG.index+1)
```

```python
## Show Monthly AVG returns 

sp500_AVG * 100
```

```python
### Plot the Average monthly returns 

sp500_AVG.plot.bar()
```

Worst Month -Above plot defines that September month is the worst month to invest contuning with february and May having a minimal dip.
